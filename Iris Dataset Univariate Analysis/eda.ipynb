```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

# 1. Categorical Data
## a. Countplot

```python
sns.countplot(x='Species',data=df)
```

```python
df.count()
```

## b. PieChart

```python
df['Species'].value_counts().plot(kind='pie',autopct='%.2f')
```

# Numerical Data
## a. Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['SepalLengthCm'],bins=50)
```

## b.Displot

```python
sns.histplot(df['SepalLengthCm'], kde=True)
```

## c. Boxplot

```python
sns.boxplot(x='SepalLengthCm', data=df)
```
